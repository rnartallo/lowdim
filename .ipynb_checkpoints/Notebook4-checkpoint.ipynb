```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import scipy as sp
import torch
import torch.nn as nn
import networkx as nx
import torch.optim as optim
import torch.nn.functional as F
```

```python
def LV(x,t,A,a,b):
    N = x.shape[0]
    dXdt = np.zeros(N)
    for n in range(0,N):
        dXdt[n] = x[n]*(a[n]-b[n]*x[n])
        for j in range(0,N):
            dXdt[n] -= A[n,j]*x[n]*x[j]
    return dXdt
```

```python
N=80
g = nx.barabasi_albert_graph(N,1)
A = nx.to_numpy_array(g)
T = 5
dt = 0.05
nT = int(T/dt) 
t = np.linspace(0,T,nT)
a = np.random.uniform(0.5,1.5,N)
b = np.random.uniform(0.5,1.5,N)
no_batches = 100
x = np.zeros((no_batches,nT,N))
for s in range(0,no_batches):
    if np.mod(s,10)==0:
        print(s)
    x0 = np.random.uniform(0,1,N)
    x[s,:,:] = sp.integrate.odeint(LV,x0,t,args=(A,a,b,))
x = torch.tensor(x, dtype=torch.float32)
```

# Architecture

```python
class Encoder(nn.Module):
    def __init__(self, input_dim, hidden_dim, latent_dim, num_layers=1):
        super(Encoder, self).__init__()
        self.lstm = nn.LSTM(input_dim, hidden_dim, num_layers, batch_first=True)
        self.fc_mu = nn.Linear(hidden_dim, latent_dim)
        self.fc_logvar = nn.Linear(hidden_dim, latent_dim)
    
    def forward(self, x):
        h, _ = self.lstm(x)
        h = h[:, -1, :]  # Get the last hidden state
        mu = self.fc_mu(h)
        logvar = self.fc_logvar(h)
        return mu, logvar

class LatentDynamics(nn.Module):
    def __init__(self, latent_dim, hidden_dim, num_layers=1):
        super(LatentDynamics, self).__init__()
        self.lstm = nn.LSTM(latent_dim, hidden_dim, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_dim, latent_dim)
    
    def forward(self, z):
        h, _ = self.lstm(z)
        next_z = self.fc(h)
        return next_z

class Decoder(nn.Module):
    def __init__(self, latent_dim, hidden_dim, output_dim, num_layers=1):
        super(Decoder, self).__init__()
        self.fc = nn.Linear(latent_dim, hidden_dim)
        self.lstm = nn.LSTM(hidden_dim, hidden_dim, num_layers, batch_first=True)
        self.output_layer = nn.Linear(hidden_dim, output_dim)
    
    def forward(self, z):
        h = self.fc(z)
        h, _ = self.lstm(h)
        output = self.output_layer(h)
        return output

# Define the VAE with Latent Dynamics
class VAE_Dynamics(nn.Module):
    def __init__(self, input_dim, hidden_dim_encoder,hidden_dim_latent,hidden_dim_decoder, latent_dim, output_dim, num_layers=1):
        super(VAE_Dynamics, self).__init__()
        self.encoder = Encoder(input_dim, hidden_dim_encoder, latent_dim, num_layers)
        self.dynamics = LatentDynamics(latent_dim, hidden_dim_latent, num_layers)
        self.decoder = Decoder(latent_dim, hidden_dim_decoder, output_dim, num_layers)
    
    def reparameterize(self, mu, logvar):
        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)
        return mu + eps * std
    
    def forward(self, x):
        mu, logvar = self.encoder(x)
        z0 = self.reparameterize(mu, logvar).unsqueeze(1)
        
        # Get the latent sequence by applying the dynamics model
        z_seq = [z0]
        for t in range(1, x.size(1)):
            next_z = self.dynamics(z_seq[-1])
            z_seq.append(next_z)
        z_seq = torch.cat(z_seq, dim=1)
        
        # Reconstruct the original sequence from the latent sequence
        recon_x = self.decoder(z_seq)
        return recon_x, mu, logvar, z_seq

    def sample(self, z0, seq_len):
        z_seq = [z0]
        for t in range(1, seq_len):
            next_z = self.dynamics(z_seq[-1])
            z_seq.append(next_z)
        z_seq = torch.cat(z_seq, dim=1)
        return z_seq,self.decoder(z_seq)

# Loss function (VAE Loss)
def vae_loss(recon_x, x, mu, logvar):
    BCE = F.mse_loss(recon_x, x, reduction='sum')
    KLD = -0.5 * torch.sum(1 + logvar - mu.pow(2) - logvar.exp())
    return BCE + KLD
```

```python
input_dim = N  # Dimensionality of input time-series
hidden_dim_encoder = 120  # Hidden layer dimensionality in encoder LSTM
hidden_dim_latent = N  # Hidden layer dimensionality
latent_dim = N  # Latent space dimensionality
hidden_dim_decoder = 120
output_dim = N  # Output dimensionality (same as input for reconstruction)
```

```python
model = VAE_Dynamics(input_dim, hidden_dim_encoder,hidden_dim_latent,hidden_dim_decoder,latent_dim, output_dim, num_layers=1)
optimizer = optim.Adam(model.parameters(), lr=6e-3)
```

```python
for epoch in range(1000):
    model.train()
    optimizer.zero_grad()
    recon_x, mu, logvar, z_seq = model(x)
    loss = vae_loss(recon_x, x, mu, logvar)
    loss.backward()
    optimizer.step()
    if np.mod(epoch,100)==0:
        print(f"Epoch {epoch + 1}, Loss: {loss.item()}")
decoded_x = recon_x.detach().numpy()
```

```python
fig, ax = plt.subplots(1,2)
fig.set_size_inches(7,3)
for n in range(0,N):
    if np.mod(n,int(N/5))==0:
        ax[0].plot(t,x[1,:,n],'black')
        ax[1].plot(t,decoded_x[1,:,n],'black')
    else:
        ax[0].plot(t,x[1,:,n],'grey',alpha=0.2)
        ax[1].plot(t,decoded_x[1,:,n],'grey',alpha=0.2)
sns.despine()
plt.tight_layout()
plt.show()
```

```python
z0 = torch.randn(1,1,latent_dim)  # Initial latent state
z_seq, sampled_seq = model.sample(z0, seq_len=nT)
sampled_seq = sampled_seq.detach().numpy().reshape(nT,N)
z_seq = z_seq.detach().numpy().reshape(nT,latent_dim)
```

```python
fig, ax = plt.subplots(1,2)
fig.set_size_inches(7,3)
for n in range(0,N):
    ax[1].plot(t,sampled_seq[:,n],'grey',alpha=0.8)
for l in range(0,latent_dim):
    ax[0].plot(t,z_seq[:,l],'grey',alpha=0.8)
sns.despine()
plt.tight_layout()
plt.show()
```
