```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import scipy as sp
import networkx as nx
```

```python
def ODE(x,t,A,f,g,params):
    N = x.shape[0]
    dXdt = np.zeros(N)
    for n in range(0,N):
        dXdt[n] = f(x[n],params[:,n])
        for j in range(0,N):
            dXdt[n] += A[n,j]*g(x[n],x[j])
    return dXdt
```

```python
def LV():
    def f(x,param):
        a = param[0]
        b = param[1]
        return x*(a-b*x)
    def g(x,y):
        return(-x*y)
    return f,g

def MP():
    def f(x,param):
        a = param[0]
        b = param[1]
        return x*(a-b*x)
    def g(x,y):
        return(x*(y**2)*((1+x**2)**(-1)))
    return f,g

def MM():
    def f(x,param):
        return -x
    def g(x,y):
        return (y**2)/(1+y**2)
    return f,g

def SIS():
    def f(x,param):
        return -param[0]*x
    def g(x,y):
        return (1-x)*y
    return f,g


def WC():
    def f(x,param):
        return -x
    def g(x,y):
        return 1/(1+np.exp(-(y-1)))
    return f,g
```

```python
def weighted_ba_graph(W):
    N = W.shape[0]
    A = np.zeros((N,N))
    for n in range(0,N):
        for m in range(0,N):
            if W[n,m]==1:
                A[n,m] = np.random.uniform(low=-1,high=1)
    return(A)
```

```python
N=50
g = nx.barabasi_albert_graph(N,1)
W = nx.to_numpy_array(g)
A = weighted_ba_graph(W)

a = np.random.uniform(0.5,1.5,N)
b = np.random.uniform(0.5,1.5,N)
LV_params = np.zeros((2,N))
LV_params[0,:] = a
LV_params[1,:] = b
MP_params = np.zeros((2,N))
MP_params[0,:] = a
MP_params[1,:] = b

MM_params = np.zeros((1,N))

SIS_params = np.random.uniform(0.5,1.5,N).reshape(1,N)
WC_params = np.zeros((1,N))

x0 = np.random.uniform(0,1,N)
LV_IC = x0
MP_IC = x0
MM_IC = np.random.uniform(0,2,N)
SIS_IC = np.random.uniform(0,0.1,N)
WC_IC = np.random.uniform(0,10,N)
```

```python
#x0 = np.random.uniform(0,1,N)
T = 5
dt = 0.01
nT = int(T/dt) 
t = np.linspace(0,T,nT)
```

```python
fig,ax = plt.subplots(1,5)
fig.set_size_inches(25,5)
sns.set(font_scale = 2, style = 'white')

models = [LV,MP,MM,SIS,WC]
models_params = [LV_params,MP_params,MM_params,SIS_params,WC_params]
models_IC = [LV_IC,MP_IC,MM_IC,SIS_IC,WC_IC]
models_names = ['LV','MP','MM','SIS','WC']

for idx,model in enumerate(models):
    f,g = model()
    x = sp.integrate.odeint(ODE,models_IC[idx],t,args=(A,f,g,models_params[idx],))
    #for n in range(0,int(N/5)):
        #ax[idx].plot(t,x[:,n*1],'black')
    for n in range(0,N):
        ax[idx].plot(t,x[:,n],'grey',alpha=0.5)
    ax[idx].set_title(models_names[idx])

sns.despine()
plt.tight_layout()
plt.show()
```

## Network inference

```python
def derivative_approx(x,dt):
    dx = np.zeros(x.shape)
    for i in range(0,x.shape[1]):
        for t in range(0,x.shape[0]-1):
            dx[t,i] = (x[t+1,i]-x[t,i])*(1/dt)
    return(dx)
```

```python
f,g = LV()
x = sp.integrate.odeint(ODE,LV_IC,t,args=(A,f,g,LV_params,))
dx = derivative_approx(x,dt)
Y = dx-f(x,LV_params)
```

```python
A_particular = np.zeros((N,N))
reg = 1
for i in range(0,N):
    Y_i = Y[:,i]
    G_i = np.zeros((nT,N))
    for k in range(0,nT):
        for j in range(0,N):
            G_i[k,j] = g(x[k,i],x[k,j])
    #a_i_p = np.linalg.pinv(G_i)@Y_i
    a_i_p = np.linalg.inv(G_i.T@G_i + reg*np.eye(N))@(G_i.T)@Y_i
    A_particular[i,:] = a_i_p
fig,ax = plt.subplots(1,2)
fig.set_size_inches(7,3)
sns.set(font_scale = 0.5, style = 'white')
sns.heatmap(A,ax=ax[0],vmin=-1,vmax=1,cmap='mako_r')
sns.heatmap(A_particular,ax=ax[1],vmin=-1,vmax=1,cmap='mako_r')
ax[0].invert_yaxis()
ax[1].invert_yaxis()

plt.tight_layout()
plt.show()
```

```python
x_surrogate = sp.integrate.odeint(ODE,LV_IC,t,args=(A_particular,f,g,LV_params,))
```

```python
fig,ax = plt.subplots(2,2)
fig.set_size_inches(7,6)
sns.set(font_scale = 1, style = 'white')
sns.heatmap(A,ax=ax[0,0],vmin=-1,vmax=1,cmap='mako_r')
sns.heatmap(A_particular,ax=ax[0,1],vmin=-1,vmax=1,cmap='mako_r')
ax[0,0].invert_yaxis()
ax[0,1].invert_yaxis()
sns.set(font_scale = 1, style = 'white')
for n in range(0,N):
    ax[1,0].plot(t,x[:,n],'grey',alpha=0.5)
    ax[1,1].plot(t,x_surrogate[:,n],'grey',alpha=0.5)
ax[0,0].set_title(r'$A$')
ax[0,1].set_title(r'$a_{\text{p}}$')
ax[1,0].set_xlabel(r'$t$')
ax[1,1].set_xlabel(r'$t$')
ax[1,0].set_ylabel(r'Trajectories with $A$')
ax[1,1].set_ylabel(r'Trajectories with $a_{\text{p}}$')
sns.despine()
plt.tight_layout()
plt.show()
```
